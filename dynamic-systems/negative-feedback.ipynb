```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
from sympy.plotting import plot
```

```python
# define symbols
x, y = symbols('x y')
```

```python
k3 = 2
k4 = 2
```

```python
v1 = k3 * x
v2 = k4
```

```python
# try J1, J2 from 0.1 to 1.0 in 0.1 increments, and plot the results on the same graph
J1 = np.arange(0.1, 1.1, 0.1)
J2 = np.arange(0.1, 1.1, 0.1)

sols = []
legend = []

for j1 in J1:
    j2 = j1
    lhs = v1 * (1-y)/(j1 + 1-y)
    rhs = v2*y/(j2 + y)
    equation = Eq(lhs, rhs)
    sol = solve(equation, y)
    sols.append(sol[1])
    legend.append(f'J1= {j1:.2f}, J2= {j2:.2f}')

# plot
pic = plot(*sols, (x, 0, 2), ylim=(0, 1), ylabel='y', xlabel='x', show=False)
for p in pic:
    p.label = legend.pop(0)
pic.legend = True
pic.show()
```
